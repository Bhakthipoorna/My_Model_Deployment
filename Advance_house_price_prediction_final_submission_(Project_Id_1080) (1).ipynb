##Looking for a features which has strong correlation
#####1.Dropping the features which have strong correlation

```python
def correlation(dataset,threshold):
  col_corr=set()
  corr_matrix=dataset.corr()
  for i in range (len(corr_matrix.columns)):
    for j in range (i):
      if (corr_matrix.iloc[i,j])>=threshold:
        col_name=corr_matrix.columns[i]
        col_corr.add(col_name)
    return col_corr
```

```python
corr_features=correlation(new_df,0.7)
len(corr_features)
```

## Removing outliers using Zscore

```python
from  scipy import stats
z=np.abs(stats.zscore(new_df))
z

threshold=4
print(np.where(z>4))

new_dff=new_df[(z<4).all(axis=1)]
new_dff
```

##Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit_transform(new_dff)
```

## Splitting the Dependent and Independent variable

```python
x=new_dff.drop(['SalePrice'],axis=1)
y=new_dff['SalePrice']
y
```

##Splitting the data into Training and Testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Applying the model 

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

##Predicting the model

```python
y_pred=model.predict(x_test)
```

##Checking the accuracy of the model using R2_score AND ADJUSTED R2_SCORE

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print('R2_SCORE :',r2_score(y_test,y_pred)*100)
print('ADJUSTED R2_SCORE :',(1 - (1-model.score(x,y))*(len(y)-1)/(len(y)-x.shape[1]-1))*100)
```

##Checking whether the model is overfit or not

```python
print('MODEL ACCURACY IN TRAINING PHASE:',model.score(x_train,y_train)*100)
print('MODEL ACCURACY IN TESTING PHASE:',model.score(x_test,y_test)*100)
```

##CONCLUSION
